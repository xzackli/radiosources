```python
import healpy as hp

m_cib = hp.read_map("/tigress/zequnl/websky/cib_nu0100.fits", 0)
m_tsz = hp.read_map("/tigress/zequnl/websky/tsz.fits", 0);
```

```python
from astropy import units as u
from astropy.cosmology import Planck15
freq = 100 * u.GHz
equiv = u.thermodynamic_temperature(freq, Planck15.Tcmb0)
mjyk = (u.MJy / u.sr).to(u.K, equivalencies=equiv)  
```

```python
m_cib *= mjyk * 1e6
m_tsz *= -4.1103e6
```

```python
cl_cib = hp.anafast(m_cib)
cl_tsz = hp.anafast(m_tsz)
```

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
%config InlineBackend.figure_format = 'retina'
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure()

lsq = np.arange(len(cl_cib))**2 / (2 * np.pi)
plt.plot(cl_cib * lsq)
plt.plot(cl_tsz * lsq)
plt.yscale("log")
plt.ylim(1e-2, 1e3)
```
